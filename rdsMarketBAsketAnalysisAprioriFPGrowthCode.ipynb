# Apriori Market Basket Analysis with Transaction Encoder, Apriori and FPGrowth

# Import Libraries

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpgrowth
```

# Load the Dataset

```python
dataset = pd.read_csv('groceriessmall.csv',header=None,sep = ',', names=range(20))
print((dataset))
```

```python
dataset = dataset.apply(lambda x: list(x.dropna().values), axis=1).tolist()
print(dataset)
```

# Create a crosstab with transaction encoder for all the items

```python
len(dataset)
tr = TransactionEncoder()
trenc = tr.fit(dataset)
trenc = trenc.transform(dataset)
print(trenc)
dataset1 = pd.DataFrame(trenc,columns=tr.columns_)
dataset1
```

```python
apriori(dataset1,min_support=0.4,use_colnames=True)
```

```python
fpgrowth(dataset1,min_support=0.4, use_colnames=True)
```
